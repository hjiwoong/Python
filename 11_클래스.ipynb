### 클래스

* 클래스의 선언 및 인스턴스 생성
```
class Starbucks:
    brand = "Starbucks"  # 클래스 변수 (모두가 공유)

    def __init__(self, name, price): # 생성자
        self.name = name           # 인스턴스 변수 (개별적)
        self.price = price         # 인스턴스 변수 (개별적)

    def order(self):               # 메서드
        print(f"{self.brand}에서 {self.name}을(를) 주문합니다.")

# americano는 객체입니다.
# 또한 americano는 Starbucks 클래스의 인스턴스입니다.
americano = Starbucks("아메리카노", 4500)
americano.order()
```
1. 인스턴스 (Instance): 클래스(설계도)를 바탕으로 실제 메모리에 만들어진 실체, '스타벅스 메뉴'라는 설계도로 찍어낸 실제 아메리카노가 인스턴스

2. 메서드 (Method): 클래스 내부에 정의된 함수. 객체가 할 수 있는 '행동'을 정의, 커피 객체의 '주문하기'같은 기능

3. 애트리뷰트 (Attribute): 객체가 가지고 있는 상태나 성질. (인스턴스 변수와 클래스 변수를 모두 포함하는 상위 개념), 커피의 '이름', '가격'같은 정보

4. self: 객체 자기 자신을 가리키는 키워드. 메서드를 호출할 때 어떤 객체가 부른 것인지 구분하기 위해 사용

5. 생성자 (Constructor): 객체가 생성될 때 자동으로 호출되는 특수 메서드(__init__). 주로 초기값을 설정할 때 사용, 커피가 만들어지자마자 컵에 이름표를 붙이고 기본 가격을 매기는 과정

6. 인스턴스 변수 (Instance Variable): 각 인스턴스마다 개별적으로 가지는 변수. self.변수명 형태로 정의

7. 클래스 변수 (Class Variable): 해당 클래스로 만든 모든 인스턴스가 공유하는 변수. 클래스 내부, 메서드 외부에서 정의.


### 매직 메서드 Magic method


```
 __init__() : 생성자의 역할 수행, 인스턴스가 처음 생성될 때부터 필요한 변수 선언 및 메서드 호출을 실행, return 키워드 사용 불가.
 __call__() : 인스턴스를 함수처럼 호출했을 때 실행되는 메서드
 __len__() : len()함수의 입력으로 클래스의 인스턴스를 사용할때 호출되는 함수, self 이외의 인자를 받을 수 없다.
 __getitem__() : 인스턴스에 인덱싱과 슬라이싱을 수행할때 실행되는 함수, self 이외에 하나의 인자를 받고, 대괄호에 사용할 인덱싱, 슬라이싱 정보가 된다.
 __add__() : 덧셈연산자+를 사용하면 자동으로 __add__() 메서드 호출
```



###  상속 inheritance
* 객체 지향 프로그래밍의 기본 개념
* 코드의 재사용, 기능의 확장성

class 자식클래스(부모클래스):

### 오버라이딩 overriding
* 상속 받은 부모클래스의 메서드를 자식클래스에서 재정의하는 것, 부모클래스의 기능을 덮어쓰거나 확장하여 사용

### @property : 클래스의 인스턴스 변수에 접근할때 해당 메서드가 실행된다.


```
class 클래스명:
  @property
  def 변수(self):
    실행코드
  @변수.setter
  def 변수(self, 입력값):
    실행코드
```




### @staticmethod 데코레이터를 사용하면 실행 과정에서 인스턴스 변수 값이나 클래스 변수 값이 변하지 않는 정적 메서드를 선언할 수 있다.


```
class 클래스명:
  @staticmethod
  def 메서드(인자):
    실행코드
```



### @classmethod 데코레이터는 메서드를 클래스 메서드로 선언하며, 클래스 변수 값을 변경하거나 읽을때 사용


```
class 클래스명:
  @classmethod
  def 메서드(cls, 변수): # 첫번째 인자가 클래스를 가리키며 클래스변수에 접근하는데 사용
    실행 코드
```



````python
class Korean:
  country = 'korea' #클래스 변수
  @classmethod
  def change_country(cls, name): #cls는 클래스
    cls.country = name

k1=Korean()
k2=Korean()
k1.change_country('South Korea')
print(k1.country) #변경된 클래스 변수는 동일한 클래스의 다른 인스턴스에도 영향을 준다
print(k2.country)
````

````python
class Member:
  language='python'
  def __init__(self, name):
    self.name=name

  @staticmethod
  def print_upper(name):
    print(name.upper())

  @classmethod
  def change_lang(cls, lang):
    cls.language=lang

m1,m2 = Member('kim'), Member('lee')
m1.print_upper(m1.name)
m1.change_lang('PYTHON')
print(m2.language)
````

### getattr() 객체의 애트리뷰트를 문자열로 접근할때 사용하는 파이썬 내장 함수


```
getattr(객체, 이름[, 기본값])

class Member:
  def __init__(self, name):
    self.name = name

m = Member('kim')
print(getattr(m, 'name')) #m.name을 리턴한다. 존재하지 않으면 에러 발생, 기본값이 있을경우 기본값을 리턴한다.
```



````python
class XY:
  def __init__(self,x,y):
    self.x = x
    self.y = y

c=XY(3,4)
print(getattr(c, 'x'))
print(getattr(c, 'z', 'no attribute'))
````

### setattr() 클래스 객체의 애트리뷰트를 동적으로 할당할때 사용하는 파이썬 내장 함수


```
setattr(객체, 이름, 값)

class Member:
  def __init__(self, name):
    self,name = name
    
m = Member('kim')
setattr(m, 'name', 'lee') # setattr() 함수의 리턴값은 None 이다.
```



````python
class XY:
  def __init__(self,x,y):
    self.x=x
    self.y=y

c=XY(3,4)
setattr(c, 'z', 8) #인스턴스 변수 z에 8을 할당
print(c.z)
````

 인스턴스의 애트리뷰트가 존재하지 않을때 호출되는 매직 메서드이다.

````python
class XY:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __getattr__(self,n):
    print(f'no attribute {n}')

c=XY(3,4)
print(c.z)
print(c.x)
````



```
 getattr(obj, name): 객체의 속성 값을 가져오는 내장 함수입니다. (obj.name과 결과가 같다),
 속성이 없으면 에러(AttributeError) 발생

 __getattr__(self, name): 객체에 찾는 속성이 없을 때만 마지막 수단으로 호출되는 매직 메서드,
 에러가 나기 직전에 실행되어 에러를 막을 수 있음
```



```
구분    getattr() (함수)                    __getattr__ (매직 메서드)
성격    외부에서 값을 읽어오는 도구               값이 없을 때를 대비한 방어 기제
실행    시점개발자가 코드에서 직접 호출할 때        obj.abc를 찾았는데 abc가 메모리에 없을 때 자동 실행
특이사항 속성이 없으면 에러(AttributeError) 발생  에러가 나기 직전에 실행되어 에러를 막을 수 있음
```



````python
class  Starbucks:
  def __init__(self):
    self.name = '아메리카노'

coffee = Starbucks()
attr_name = input('확인할 속성명을 입력하세요 ') #사용자가 'name' 입력
#print(coffee.attr_name) #X
#getattr()을 쓰면 변수에 담긴 문자열로 접근 가능
print(getattr(coffee, attr_name)) #getattr(coffee, 'name')
#getattr() 함수 : 동적으로 값 가져오기. 사용자로부터 입력받은 문자열을 이용해 속성 값을 가져와야 할 때 사용
````

````python
class SmartStarbucks:
  def __init__(self):
    self.name = '아메리카노'

  def __getattr__(self,name): #찾는 속성이 없을 때만 실행됨
    return f"'{name}' 속성은 준비되지 않았습니다. 다른 메뉴를 찾아보세요"

coffee = SmartStarbucks()
print(coffee.name)
print(coffee.price)
````



```
setattr(obj, name, value): 객체의 속성을 설정하기 위해 사용하는 내장 함수 (obj.name = value와 결과가 같다)

__setattr__(self, name, value): 객체의 속성이 설정될 때 내부적으로 자동 호출되는 매직 메서드. 속성 설정 방식을
커스텀(재정의)하고 싶을 때 사용.
```



```
구분	    setattr() (함수)	                  __setattr__ (매직 메서드)
성격	    외부에서 속성을 조작하는 도구	            클래스 내부에 정의된 동작 규칙
사용 시점	 속성 이름을 문자열로 가지고 있을 때	      속성이 저장되는 방식을 통제하고 싶을 때
작동 방식	 내부적으로 결국 __setattr__을 실행함	    모든 속성 설정(obj.a = 1) 시 호출됨
```
