```python
#private ì†ì„± ê°’ì„ ì–»ìœ¼ë ¤ë©´ ì†ì„± ê°’ì„ ë°˜í™˜í•˜ëŠ” ë©”ì„œë“œë¥¼ ì´ìš©í•´ì•¼ í•¨
class Account:
  def __init__(self, amount=0):
    self.__balance = amount

  def deposit(self, amount):
    self.__balance += amount
    return self.__balance

  def withdraw(self, amount):
    self.__balance -= amount
    return self.__balance

  def get_balance(self):
    return self.__balance

a1=Account(50)
print(a1.get_balance())
```

```python
#í´ë˜ìŠ¤ Circle ë§Œë“¤ê¸°
#ë°˜ì§€ë¦„ ì†ì„± __radius
#ë°˜ì§€ë¦„ ì†ì„±ì„ í™•ì¸í•˜ëŠ” ë©”ì„œë“œ get_radius ë©”ì„œë“œ
#ë°˜ì§€ë¦„ ì„¤ì •í•˜ëŠ” set_radius ë©”ì„œë“œ
#ì›ì˜ ë„“ì´ êµ¬í•˜ëŠ” get_area ë©”ì„œë“œ
#ì›ì˜ ë‘˜ë ˆ êµ¬í•˜ëŠ” get_circumference ë©”ì„œë“œ
import math
class Circle:
  def __init__(self, radius=10):
    self.__radius = radius

  def get_radius(self):
    return self.__radius

  def set_radius(self, radius):
    self.__radius = radius

  def get_area(self):
    return math.pi*self.__radius**2

  def get_circumference(self):
    return 2*math.pi*self.__radius

c1=Circle(7)
print(f'ë°˜ì§€ë¦„: {c1.get_radius()}')
c1.set_radius(15)
print(f'ë°˜ì§€ë¦„: {c1.get_radius()}')
print(f'ë„“ì´: {c1.get_area():.2f}')
print(f'ë‘˜ë ˆ: {c1.get_circumference():.2f}')
```

## 3. ìƒì† (Inheritance)
* **ê°œë…**: ê¸°ì¡´ í´ë˜ìŠ¤(ë¶€ëª¨ í´ë˜ìŠ¤)ì˜ ì†ì„±ê³¼ ë©”ì„œë“œë¥¼ ê·¸ëŒ€ë¡œ ë¬¼ë ¤ë°›ì•„ ìƒˆë¡œìš´ í´ë˜ìŠ¤(ìì‹ í´ë˜ìŠ¤)ë¥¼ ë§Œë“œëŠ” ê¸°ëŠ¥ì…ë‹ˆë‹¤.
* **ì¥ì **: ì½”ë“œì˜ ì¬ì‚¬ìš©ì„±ì„ ë†’ì—¬ì£¼ë©°, ì¤‘ë³µëœ ì½”ë“œë¥¼ ì¤„ì—¬ ìœ ì§€ë³´ìˆ˜ê°€ ì‰¬ì›Œì§‘ë‹ˆë‹¤.
* **ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©(Overriding)**: ë¶€ëª¨ í´ë˜ìŠ¤ë¡œë¶€í„° ë¬¼ë ¤ë°›ì€ ë©”ì„œë“œë¥¼ ìì‹ í´ë˜ìŠ¤ì—ì„œ ìš©ë„ì— ë§ê²Œ ìƒˆë¡­ê²Œ ì¬ì •ì˜í•˜ì—¬ ì‚¬ìš©í•˜ëŠ” ê²ƒì„ ë§í•©ë‹ˆë‹¤.

```python
#ë¶€ëª¨ í´ë˜ìŠ¤
class Student:
  def __init__(self, name, computer):
    self.name = name
    self.computer = computer

  def set_computer(self, computer):
    self.computer = computer
  def get_name(self):
    return self.name

#ìì‹ í´ë˜ìŠ¤
class Science(Student):
  def __init__(self, name, computer, science):
    super().__init__(name, computer)
    self.science=science

  def set_science(self, science):
    self.science = science
  def get_average(self):
    return(self.science + self.computer)/2

#ìì‹ í´ë˜ìŠ¤
class Social(Student):
  def __init__(self, name, computer, social):
    super().__init__(name, computer)
    self.social = social
  def set_social(self, social):
    self.social=social
  def get_average(self):
    return(self.social + self.computer)/2
```

```python
st1 = Science('lee', 90, 80)
print(st1.get_name())
print(st1.get_average())
st1.set_computer(100)
print(st1.get_average())
```

```python
st2=Social('kim', 98, 88)
print(st2.get_name())
print(st2.get_average())
st2.set_social(96)
print(st2.get_average())
```

```python
#ë¶€ëª¨ í´ë˜ìŠ¤
class Figure:
  def __init__(self, area, perimeter):
    self.area = area
    self.perimeter = perimeter

  def get_area(self):
    return self.area

  def get_perimeter(self):
    return self.perimeter

#ìì‹ í´ë˜ìŠ¤
class Circle(Figure):
  def __init__(self, radius):
    area = 3.14*radius**2
    perimeter = 2*3.14*radius
    super().__init__(area, perimeter)
    self.radius = radius

  def set_radius(self, radius):
    self.radius = radius
    self.area = 3.14*radius**2
    self.perimeter = 2*3.14*radius

  def get_radius(self):
    return self.radius

#ìì‹ í´ë˜ìŠ¤
class Rectangle(Figure):
  def __init__(self, width=10, height=10):
    area=width*height
    perimeter=2*(width+height)
    super().__init__(area, perimeter)
    self.width=width
    self.height=height

  def set_width(self, width):
    self.width=width
    self.area = width*self.height
    self.perimeter = 2*(width+self.height)

  def set_height(self, height):
    self.height=height
    self.area = self.width*height
    self.perimeter = 2*(self.width+height)

  def get_width(self):
    return self.width

  def get_height(self):
    return self.height
```

```python
f1=Circle(7)
print(f1.area)
print(f1.perimeter)
print(f1.get_area(), f1.get_perimeter())
f1.set_radius(10)
print(f1.get_area(), f1.get_perimeter())
```

```python
f2=Rectangle(5,4)
print(f2.area)
print(f2.perimeter)
print(f2.get_area(), f2.get_perimeter())
f2.set_width(10)
print(f2.get_area(), f2.get_perimeter())
f2.set_width(15)
print(f2.get_area(), f2.get_perimeter())
```
