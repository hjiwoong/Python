### 함수의 scope
* 전역 변수: 모듈(파일) 내에서 접근할 수 있으며 모듈 내 어디에서든지 접근할 수 있다.
* 지역 변수: 함수 내에서 정의된 변수이며, 해당 함수 내에서만 접근 할 수 있다. 해당 함수의 실행이 종료되면 더 이상 함수 안의 지역 변수에 접근할 수 없다.

````python
#파이썬에서는 같은 변수 이름이 다른 값을 가질 수 있다
num=41 #전역변수
def number():
  num=3 #지역변수
  print(f'함수 안에서, num = {num}')
number()
print(f'global, num = {num}')
````

````python
num=41 #전역 변수
def number():
  global num #함수 내에서 전역 변수로 선언
  num = 3 #전역 변수값 변경
  print(f'함수 안에서, num = {num}')
number()
print(f'global, num = {num}')
````

### 재귀 함수 : 자기 자신을 호출하는 함수

````python
def factorial(n):
  if n ==1:
    return 1
  else:
    return n * factorial(n-1)

print(factorial(3))
````

### 가변 인자

*args : 순차적인 입력에 대해서 개수에 상관없이 입력을 받을 수 있다.
```
def 함수(*args):
  실행코드

```

**kwargs : 키워드 인자에 대해서 어떤 이름의 인자 입력이든 자유롭게 입력 받을 수 있다.
```
def 함수(**kwargs):
  실행코드
```

func(*args, **kwargs) : 함수 호출시 어떤 형태의 입력이든 받을 수 있다.


````python
def mul(*args):
  result = 1
  for i in args:
    result *= i
  return result
print(mul(3,4))
print(mul(5))
print(mul(2,2,2,2,2))
````

````python
def group_by_age(**kwargs):
  group = {'adult': [], 'non-adult': []}
  for name, age in kwargs.items(): #딕셔너리.items()
    if age > 18:
      group['adult']+=[name]
    else:
      group['non-adult']+=[name]
  return group
print(group_by_age())
print(group_by_age(kim=25, lee=16, park=30))
````

````python
def free_arg(*args, **kwargs): #순서 중요
  print(f'args = {args}') #args는 튜플
  print(f'kwargs = {kwargs}') #kwargs는 딕셔너리

free_arg()
free_arg(1,2)
free_arg(first='1st', second='2nd')
free_arg(5, first='1st')
````

### lambda 문법 : 함수를 선언하는 문법, 한줄로 간단하게 정의되는 함수를 만들때 사용, 특정 입력 값을 고정한 새 함수를 만들때도 사용



```
lambda x, y : 리턴값
```



````python
def mul(x,y):
  return x*y
print(mul(2,3))
````

````python
nums=[5,0,-7,-10]
print(sorted(nums,key=lambda x: x**2))
````

````python
eq = lambda a, b, c: f'{a}x + {b}y = {c}'
print(eq(5,4,3))
````

````python
slash_join=lambda x: '/'.join(x)
print(slash_join(['apple', 'banana', 'grape']))
````

### 함수의 객체화
* 함수를 변수에 대입하고 다른함수의 인자로 입력


````python
def mul(a,b):
  return a*b

calculate=mul
print(calculate(3,5))
````

````python
def function(fnc):
  print(f'fnc(2,3) = {fnc(2,3)}')

def mul(a,b):
  return a*b

function(mul)
````

````python
def function():
  def new_function(a,b):
    return a*b
  return new_function
mul = function()
print(mul(2,3))
````

````python
def double(x):
  return 2*x

def func_with_print(fnc):
  def new_func(x):
    y=fnc(x)
    print(f'y={y}')
    return y
  return new_func

new_double=func_with_print(double)
print(new_double(5))
````

### 데코레이터 decorator : 함수의 기능을 변경, 확장할 수 있는 문법
* 기존 함수의 기능은 수정하지 않고 함수 실행의 앞뒤에 미리 정의한 코드를 실행 할 수 있다.
* 원본 함수를 입력받아, 기능을 추가한 새로운 함수를 만들어 반환하는 함수


```
@데코레이터
def 함수명(인자):
  실행코드
```



````python
def deco(fnc):
  def new_fnc():
    print('fnc() is called')
    return fnc()
  return new_fnc

@deco
def hi():
  print('hi')

hi()
````

````python
def deco(fnc):
  def new_fnc():
    print('fnc() is called')
    return fnc()
  return new_fnc

def hi():
  print('hi')

hi=deco(hi) #데코레이터 함수에 인자값으로 전달되고 다시 리턴된다. 위와 동일한 동작
    #hi() 함수는 놔두고 더 확장된기능을 추가할 수 있다
hi()
````

````python
def func_with_print(fnc): #원본 함수를 인자로 입력 받는다. 그리고 기능을 추가한 새로운 함수를 만들어서 반환
  def new_func(x):
    y=fnc(x)
    print(f'y={y}')
    return y
  return new_func

@func_with_print
def double(x): #원본 함수
  return 2*x

print(double(5))
````

### 클로저 closure : 함수 안 지역 변수를 함수 밖에서도 간접적으로 사용할 수 있다.

````python
def function(a):
  def inner_function(b):
    return a*b
  return inner_function

double = function(2) #double = inner_function, a=2
triple = function(3) #triple = inner_function, a =3
print(double(10)) #inner_function(10), b=10, 외부 함수로부터 물려받은 a와 자신이 받은 인자 b를 곱한 값을 반환
print(triple(10)) #inner_function(10), b=10, 자기가 정의된 구역 밖에서 호출되어도 자기를 만들어준 외부 함수의 변수 a에 접근할 수 있고 이를 '자유변수(Free variable)를 캡쳐한다'고 표현한다
````

````python
#매번 똑같은 머리말(로그 레벨, 이름 등)을 붙여야 할 때 클로저 사용
def make_logger(level):
  def log(message):
    print(f'[{level}]{message}')
  return log

info_logger = make_logger('INFO')
error_logger = make_logger('ERROR')

info_logger('서버가 시작되었습니다') #클로저는 함수가 종료된 후에도 내부의 지역 변수를 기억. 전역변수처럼 값이 유지
info_logger('서버가 종료되었습니다')
error_logger('데이터베이스 연결 실패')
````
