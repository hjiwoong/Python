## 1. ë¹„êµ ì—°ì‚°ì (Comparison Operators)
* **ê°œë…**: ë‘ ê°’ì˜ í¬ê¸°ë¥¼ ë¹„êµí•˜ì—¬ ì°¸(`True`) ë˜ëŠ” ê±°ì§“(`False`)ì„ ê²°ê³¼ë¡œ ë°˜í™˜í•˜ëŠ” ì—°ì‚°ìì…ë‹ˆë‹¤.
* **ì£¼ìš” ì—°ì‚°ì**:
    * `==` : ê°™ë‹¤
    * `!=` : ë‹¤ë¥´ë‹¤
    * `>` / `<` : í¬ë‹¤ / ì‘ë‹¤
    * `>=` / `<=` : í¬ê±°ë‚˜ ê°™ë‹¤ / ì‘ê±°ë‚˜ ê°™ë‹¤

````python
a, b, c = 10, 20, 30
print(f'{a}=={b}: {a==b}')
print(f'{a}!={b}: {a!=b}')
print(f'{a}>{b}: {a>b}')
print(f'{b}<{c}: {b<c}')
print(f'{a}>={b}: {a>=b}')
print(f'{b}<={c}: {b<=c}')
````

### ğŸ“ ë‘ ë„í˜•ì˜ ë„“ì´ ë¹„êµí•˜ê¸°
1. **ê³„ì‚°**: ì‚¬ê°í˜•(ê°€ë¡œÃ—ì„¸ë¡œ)ê³¼ ì‚¼ê°í˜•(ë°‘ë³€Ã—ë†’ì´/2)ì˜ ì •ë³´ë¥¼ ê°ê° ì…ë ¥ë°›ì•„ ë„“ì´ë¥¼ êµ¬í•©ë‹ˆë‹¤.
2. **ë¹„êµ**:
    * `if ë„“ì´1 > ë„“ì´2:` ì²« ë²ˆì§¸ ë„í˜•ì´ ë” ë„“ë‹¤ê³  ì¶œë ¥í•©ë‹ˆë‹¤.
    * `elif ë„“ì´1 < ë„“ì´2:` ë‘ ë²ˆì§¸ ë„í˜•ì´ ë” ë„“ë‹¤ê³  ì¶œë ¥í•©ë‹ˆë‹¤.
    * `else:` ë‘ ë„í˜•ì˜ ë„“ì´ê°€ ê°™ë‹¤ê³  ì¶œë ¥í•©ë‹ˆë‹¤.

````python
# ì‚¬ê°í˜•ì€ ê°€ë¡œì„¸ë¡œ, ì‚¼ê°í˜•ì€ ë°‘ë³€ë†’ì´, ì›ì€ ë°˜ì§€ë¦„ í•¨ìˆ˜ 3ê°œ
import math
def square():
  width=int(input('ì‚¬ê°í˜•ì˜ ê°€ë¡œ: '))
  height=int(input('ì‚¬ê°í˜•ì˜ ì„¸ë¡œ: '))
  return width*height
def triangle():
  base=int(input('ì‚¼ê°í˜•ì˜ ë°‘ë³€: '))
  height=int(input('ì‚¼ê°í˜•ì˜ ë†’ì´: '))
  return base*height/2
def circle():
  radius=int(input('ë°˜ì§€ë¦„: '))
  return math.pi*radius*radius

shape_type1=input('ë„í˜•ì˜ ì¢…ë¥˜(ì‚¬ê°í˜•, ì‚¼ê°í˜•, ì›): ')
if shape_type1=='ì‚¬ê°í˜•':
  area1=square()
  print(f'{shape_type1}ì˜ ë„“ì´: {area1}')
elif shape_type1=='ì‚¼ê°í˜•':
  area1=triangle()
  print(f'{shape_type1}ì˜ ë„“ì´: {area1}')
elif shape_type1=='ì›':
  area1=circle()
  print(f'{shape_type1}ì˜ ë„“ì´: {area1}')
else:
  print('"ì‚¬ê°í˜•", "ì‚¼ê°í˜•, "ì›" ì¤‘ ì…ë ¥')

shape_type2=input('ë„í˜•ì˜ ì¢…ë¥˜(ì‚¬ê°í˜•, ì‚¼ê°í˜•, ì›): ')
if shape_type2=='ì‚¬ê°í˜•':
  area2=square()
  print(f'{shape_type2}ì˜ ë„“ì´: {area2}')
elif shape_type2=='ì‚¼ê°í˜•':
  area2=triangle()
  print(f'{shape_type2}ì˜ ë„“ì´: {area2}')
elif shape_type2=='ì›':
  area2=circle()
  print(f'{shape_type2}ì˜ ë„“ì´: {area2}')
else:
  print('"ì‚¬ê°í˜•", "ì‚¼ê°í˜•, "ì›" ì¤‘ ì…ë ¥')

if area1>area2:
  print(f'ì²« ë²ˆì§¸ {shape_type1}ì´ {area1-area2}cmÂ²ë§Œí¼ ë” ë„“ìŠµë‹ˆë‹¤')
elif area1<area2:
  print(f'ë‘ ë²ˆì§¸ {shape_type2}ì´ {area2-area1}cmÂ²ë§Œí¼ ë” ë„“ìŠµë‹ˆë‹¤')
else:
  print(f'ë‘ ë„í˜•ì˜ ë„“ì´ëŠ” {area1}cmÂ²ë¡œ ê°™ìŠµë‹ˆë‹¤')
````
